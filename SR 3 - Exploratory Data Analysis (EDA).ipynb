```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:/Users/Admin/Favorites/Documents/TECHNICAL PROGRAMMING II/settle2.csv"
data = pd.read_csv(file_path, skiprows=4)  # Adjust skiprows if the initial rows contain metadata

# Ensure the dataset is loaded correctly
print(data.head())

# Drop rows with missing values
data.dropna(inplace=True)

# Select numerical and categorical columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Plot distribution charts for numerical columns
plt.figure(figsize=(18, 6))
for i, column in enumerate(numeric_columns):
    plt.subplot(1, len(numeric_columns), i+1)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Plot categorical columns
num_categorical_columns = len(categorical_columns)
num_rows = (num_categorical_columns + 1) // 2  # Calculate number of rows needed
plt.figure(figsize=(18, num_rows * 6))
for i, column in enumerate(categorical_columns):
    plt.subplot(num_rows, 2, i+1)
    sns.countplot(x=column, data=data)
    plt.title(f'Counts of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
