```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:/Users/Admin/Favorites/Documents/TECHNICAL PROGRAMMING II/API_EN.POP.SLUM.UR.ZS_DS2_en_csv_v2_289192/API_EN.POP.SLUM.UR.ZS_DS2_en_csv_v2_289192.csv"

# Load the dataset
data = pd.read_csv(file_path, skiprows=5)

# Print the column names
print(data.columns)

# Basic EDA
print("Summary Statistics:")
print(data.describe())

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(data['Population living in slums (% of urban population)'], bins=20, kde=True)
plt.title("Distribution of Urban Population Living in Slums")
plt.xlabel("Urban Population Living in Slums (% of Urban Population)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
non_numeric_columns = data.select_dtypes(exclude=['float', 'int']).columns
print("Non-numeric columns:", non_numeric_columns)

# Remove non-numeric columns or convert them to numeric types if possible
# Drop non-numeric columns
data_numeric = data.drop(columns=non_numeric_columns)

# correlation matrix for numeric columns
correlation_matrix = data_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
